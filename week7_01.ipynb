```python
data['discount'] = data['discount'].fillna('0')
```

```python
data
```

```python
discount_category = []
for i in range(len(data)):
    if float(data['discount'][i]) > 0.5:
        discount_category.append("high")
    else:
        discount_category.append("low")
```

```python
data['discount_category'] = discount_category
```

```python
data
```

```python
plt.figure(figsize = (7,7))
sn.barplot(data = data, x = "discount_category", y = "sales")
```

```python
plt.figure(figsize = (7,7))
x = sn.barplot(data = data, x = "discount_category", y = "sales", errwidth = 0)
x.set_xlabel("Low or High Discount", fontsize = 15)
x.set_ylabel("Sales", fontsize = 15)
```

```python
likes_category= []
for i in range(len(data)):
    if float(data['likes'][i]) > 10000:
        likes_category.append("high")
    else:
        likes_category.append("low")
```

```python
data['likes_category'] = likes_category
```

```python
plt.figure(figsize = (7,7))
x = sn.barplot(data = data, x = "likes_category", y = "discount")
x.set_xlabel("Low or High Like Count", fontsize = 15)
x.set_ylabel("Discount", fontsize = 15)
```
