```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = LogisticRegression()
result1 = model1.fit(X_train, y_train)
```

```python
model2 = LinearSVC()
result2 = model2.fit(X_train, y_train)
```

```python
model3 = DecisionTreeClassifier(max_depth = 3)
result3 = model3.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def accuracy_fun(model, x, y):
    y_pred = model.predict(x)
    accuracy = accuracy_score(y, y_pred)
    return accuracy
```

```python
model1_accuracy = accuracy_fun(model1, X_test, y_test)
print(f"Accuracy: {model1_accuracy: .5f}")
```

```python
model2_accuracy = accuracy_fun(model2, X_test, y_test)
print(f"Accuracy: {model2_accuracy: .5f}")
```

```python
model3_accuracy = accuracy_fun(model3, X_test, y_test)
print(f"Accuracy: {model3_accuracy: .5f}")
```
